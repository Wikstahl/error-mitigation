```python
import networkx as nx
import os
```

```python
# Generate 30 Erd√∂s Renyi graphs
for i in range(30):
    # Folder
    folder = "max_cut_" + str(i)

    # Create the directory
    os.mkdir(folder)
    
    # Generate Graph
    G = nx.generators.random_graphs.erdos_renyi_graph(n=6,p=0.5)
    
    path = folder + "/graph"
    # Write graph in Python pickle format.
    nx.readwrite.gpickle.write_gpickle(G, path, protocol=5)
```

```python
# This is how you read files
G = nx.readwrite.gpickle.read_gpickle('max_cut_0/graph')
nx.draw_circular(G, node_size=1000, with_labels=True)
```
